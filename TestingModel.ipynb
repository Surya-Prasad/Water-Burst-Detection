```python
import numpy as np
import keras
from keras import backend as k
from keras.models import Sequential
from keras.layers import Activation
from keras.layers.core import Dense,Flatten
from keras.optimizers import Adam
import math
from keras.metrics import categorical_crossentropy
from keras.preprocessing.image import ImageDataGenerator
from keras.layers.normalization import BatchNormalization
from keras.layers.convolutional import *
from matplotlib import pyplot as plt
import itertools
from sklearn.metrics import confusion_matrix
%matplotlib inline
```

```python
vgg16_model=keras.applications.vgg16.VGG16(weights='D:/water_burst/vgg16_weights_tf_dim_ordering_tf_kernels.h5')
```

```python
vgg16_model.summary()
```

```python
model=Sequential()
for layer in vgg16_model.layers[:-3]:
    model.add(layer)
```

```python
for layer in model.layers:
    layer.trainable=False
```

```python
model.add(Dense(1024,activation='relu'))
model.add(Dense(512,activation='relu'))
model.add(Dense(100,activation='relu'))
model.add(Dense(2,activation='softmax'))
```

```python
model.summary()
```

```python
train_path='D:/water_burst/testing/Train'
valid_path='D:/water_burst/testing/Valid'
```

```python
train_batches=ImageDataGenerator().flow_from_directory(
    train_path,
    target_size=(224,224),
    classes=['Water','Non'],
    batch_size=20
)
valid_batches=ImageDataGenerator().flow_from_directory(
    valid_path,
    target_size=(224,224),
    classes=['Water','Non'],
    batch_size=10
)
```

```python
checkpoint = keras.callbacks.ModelCheckpoint("D:/water_burst/vgg16/weights-{loss:.5f}.h5",
                                             monitor='val_loss',
                                             verbose=1,
                                             save_best_only=True,
                                             save_weights_only=True,
                                             mode='auto',
                                             period=1)

model.compile(Adam(lr=0.0001),loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit_generator(train_batches,steps_per_epoch=math.ceil(779/20),validation_data=valid_batches,validation_steps=4,epochs=100,callbacks=[checkpoint])
```

```python
model.load_weights('D:/test/vgg.h5')
```
